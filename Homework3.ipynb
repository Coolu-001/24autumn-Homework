1.编写 Python程序,判断输入a是否为质数

```python
a = int(input("请输入一个整数："))
result = True
if a == 0 :
    result = False
else :
    for i in range (2,a) :
        if a % i == 0 :
            result = False
            break
print(result)

```

2.用Python实现插入排序该算法，不允许直接调用插入排序函数，需要手动实现其逻辑过程。

```python
def InsertionSort(arr):
    for i in range(1, len(arr)):
        temp = arr[i]
        j = i - 1
        while j >= 0 and temp < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = temp
    return arr

input_str = input("请输入一个待排序的整数数组：")
int_list = list(map(int, input_str.split()))
print(InsertionSort(int_list))
```

3.用Python实现选择排序该算法，不允许直接调用插入排序函数，需要手动实现其逻辑过程。

```python
def selection_sort(arr):
    for i in range(len(arr)):
        for j in range (i+1,len(arr)):
            if arr[j]<arr[i] :
                arr[i], arr[j] = arr[j], arr[i]
    return arr
input_str = input("请输入一个待排序的整数数组：")
int_list = list(map(int, input_str.split()))
print(selection_sort(int_list))
```

4.用Python实现快速排序该算法，不允许直接调用插入排序函数，需要手动实现其逻辑过程。

```python
import random

def partition(arr, low, high):
    pivot = arr[high]
    i = low - 1  

    for j in range(low, high):
        if arr[j] <= pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]  

    arr[i + 1], arr[high] = arr[high], arr[i + 1]  
    return i + 1


def randomized_partition(arr, low, high):
    random_pivot = random.randint(low, high)
    arr[random_pivot], arr[high] = arr[high], arr[random_pivot]  
    return partition(arr, low, high)

def quick_sort(arr, low, high):
    if low < high:
        pi = randomized_partition(arr, low, high)
        quick_sort(arr, low, pi - 1)
        quick_sort(arr, pi + 1, high)
input_str = input("请输入一个待排序的整数数组：")  
int_list = list(map(int, input_str.split()))  
quick_sort(int_list, 0, len(int_list) - 1)  
print(int_list) 
```

5.使用 Python 随机生成三组长度递增的随机数列，长度分别为100、1000、10000，对生成的随机数组使用不同的排序算法，例如选择排序和归并排序，至少三种排序方法。对这些数列的数据排序，并分析不同排序算法在不同长度数列下的运行效果，例如运行时间等指标。

```python
import random
from datetime import datetime
import matplotlib.pyplot as plt
 
def generate_random_numbers(length):
    return [random.randint(0, 1000) for _ in range(length)]
 
random_numbers100 = generate_random_numbers(100)
random_numbers1000 = generate_random_numbers(1000)
random_numbers10000 = generate_random_numbers(10000)
time1 = datetime.now()
InsertionSort(random_numbers100)
time2 = datetime.now()
InsertionSort(random_numbers1000)
time3 = datetime.now()
InsertionSort(random_numbers10000)
time4 = datetime.now()
x = [100, 1000, 10000]
y_insertion = [(time2 - time1).total_seconds(), (time3 - time2).total_seconds(), (time4 - time3).total_seconds()]


time5 = datetime.now()
selection_sort(random_numbers100)
time6 = datetime.now()
selection_sort(random_numbers1000)
time7 = datetime.now()
selection_sort(random_numbers10000)
time8 = datetime.now()
x = [100, 1000, 10000]
y_selection = [(time6 - time5).total_seconds(), (time7 - time6).total_seconds(), (time8 - time7).total_seconds()]

time9 = datetime.now()
quick_sort(random_numbers100, 0, 99)
time10 = datetime.now()
quick_sort(random_numbers1000, 0, 999)
time11 = datetime.now()
quick_sort(random_numbers10000, 0, 9999)
time12 = datetime.now()
x = [100, 1000, 5000]
y_quick = [(time10 - time9).total_seconds(), (time11 - time10).total_seconds(), (time12 - time11).total_seconds()]

```

```python
# 创建折线图
plt.figure(figsize=(10, 6))  # 设置图形大小

# 绘制插入排序数据
plt.plot(x, y_insertion, marker='o', color='b', label='Insertion Sort')

# 绘制选择排序数据
plt.plot(x, y_selection, marker='*', color='r', label='Selection Sort')

plt.plot(x, y_quick, marker='o', color='g', label='Quick Sort')

# 添加标题和标签
plt.title("Sorting Algorithm Time Complexity", fontsize=16)
plt.xlabel("Size", fontsize=14)
plt.ylabel("Time (seconds)", fontsize=14)

# 设置对数坐标轴
plt.xscale('log')
plt.yscale('log')

# 显示网格
plt.grid(True)

# 添加图例
plt.legend()

# 显示图形
plt.show()
```
